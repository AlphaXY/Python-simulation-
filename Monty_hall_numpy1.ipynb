```python
#!/usr/bin/env python
# coding: utf-8




get_ipython().run_line_magic('matplotlib', 'inline')

import numpy as np
import matplotlib.pyplot as plt

portes = np.array([1, 2, 3])
# La bonne porte est (aléatoirement) l'une des trois portes 1, 2 ou 3. Si on répète l'expérience Monty Hall dix fois
# par exemple, un tableau qui modélise 10 bonnes portes associées à dix Monty Hall est donné par:
Bonnes_portes = np.random.randint(1, 4, 10)
# Même philosophie pour le premier choix d'un participant qui est aussi aléatoire. Un tableau qui modélise 10 premiers
#choix associées à 10 Monty Hall est donné par:
Premiers_choix = np.random.randint(1, 4, 10)
# La stratégie gagnante est celle qui donne à l'issue de chaque expérience (deuxième choix == bonne porte).
# Si le participant suit la stratégie (garder son premier choix), alors dans ce cas (le deuxième choix == le premier choix)
# Dans ce cas là, il suffit de comparer les deux tableaux Bonnes_portes et Premier_choix pour savoir le nombre des gains
#d'un participant sur 10  s'il choisit la stratégie garder, le tableau suivant à dix accés 
#affiche True s'il gagne et False si non

gains_garder = Bonnes_portes == Premiers_choix

# On effectue les transformations suivantes pour obtenir un tableau qui affiche 1 si le participant gagne et 0 si non
# on remplace True par 1
gains_garder = np.where(gains_garder==True, 1, gains_garder)
# on remplace False par 0
gains_garder = np.where(gains_garder==False, 0, gains_garder)

print(gains_garder)

# le nombre des gains si le participant décide de choisir la stratégie garder son premier choix
#durant dix expériences est

Nbre_gains_garder = np.sum(gains_garder)
print("le nombre des gains si le participant décide de choisir la stratégie garder est {} sur dix".format(np.sum(gains_garder)))

# Le cas (gains_garder = 0) est celle qui correspond à stratégie 
# garder son premier choix est la stratégie perdante. Dans ce cas, le participant a dû 
# changer son choix pour qu'il obtienne (deuxième choix == bonne porte) à l'issue de l'expérience.
# Donc, un tableau associé à un participant suivant la stratégie changer son premier choix 
#qui affiche True s'il  gagne et False si non est donné par

gains_changer = Bonnes_portes != Premiers_choix

# On effectue les transformations suivantes pour obtenir un tableau qui affiche 1 si le participant gagne et 0 si non
# on remplace True par 1
gains_changer = np.where(gains_changer==True, 1, gains_changer)
# on remplace False par 0
gains_changer = np.where(gains_changer==False, 0, gains_changer)

print(gains_changer)

# le nombre des gains si le participant décide de choisir la stratégie changer son premier choix
#durant dix expériences est

Nbre_gains_changer = np.sum(gains_changer)
print("le nombre des gains si le participant décide de choisir la stratégie changer est {} sur dix".format(np.sum(gains_changer)))



figure = plt.figure()

plot_1=plt.plot(gains_changer)
plt.title("plot: dix Monty Hall")

# In[19]:


Bonnes_portes_2 = np.random.randint(1, 4, 10000)

Premiers_choix_2 = np.random.randint(1, 4, 10000)


gains_garder_2 = Bonnes_portes_2 == Premiers_choix_2


gains_garder_2 = np.where(gains_garder_2==True, 1, gains_garder_2)

gains_garder_2 = np.where(gains_garder_2==False, 0, gains_garder_2)

Nbre_gains_garder_2 = np.sum(gains_garder_2)

gains_changer_2 = Bonnes_portes_2 != Premiers_choix_2


gains_changer_2 = np.where(gains_changer_2==True, 1, gains_changer_2)

gains_changer_2 = np.where(gains_changer_2==False, 0, gains_changer_2)



Nbre_gains_changer_2 = np.sum(gains_changer_2)
figure = plt.figure()
plot_2=plt.plot(gains_changer_2)
plt.title("plot: dix mille Monty Hall")

# In[21]:

figure = plt.figure()
plot_scat = plt.scatter(range(10000), gains_changer_2)
plt.title("scatter: dix mille Monty Hall")

# In[22]:

figure = plt.figure()
plot_bars = plt.bar([1,2], [np.sum(gains_changer_2), np.sum(gains_garder_2)],tick_label=["changer", "garder"])


# In[57]:


samples =  [1000, 10000, 20000, 50000, 80000, 100000]
gain_changes =np.array([])
gain_gards = np.array([])
# La boucle for ici est nécessaire pour générer le graphique
for i in samples:
    Bonnes_portes = np.random.randint(1, 4, i)

    Premiers_choix = np.random.randint(1, 4, i)
    
    gains_garder = Bonnes_portes == Premiers_choix


    gains_garder = np.where(gains_garder==True, 1, gains_garder)

    gains_garder = np.where(gains_garder==False, 0, gains_garder)
    
    
    gains_changer = Bonnes_portes != Premiers_choix
    

    gains_changer = np.where(gains_changer==True, 1, gains_changer)

    gains_changer = np.where(gains_changer==False, 0, gains_changer)
    gain_gards= np.append(gain_gards, np.sum(gains_garder))
    gain_changes= np.append(gain_changes, np.sum(gains_changer))
    


figure = plt.figure()
plot_1 = plt.scatter(samples,gain_changes)
plot_2 = plt.scatter(samples,gain_gards)




```
